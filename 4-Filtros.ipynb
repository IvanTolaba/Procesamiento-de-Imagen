```python
def barlett(image):
  import numpy as np
  import cv2 as cv
  from matplotlib import pyplot as plt
  kernel = np.array ([[1, 2, 1],     # Este es el filtro establecido, 
                [ 2, 4, 2],           #que es el núcleo de convolución,funciona para mas dimensiones
                [  1, 2, 1]])
  dst = cv.filter2D(image,-1,kernel) #cv lo normaliza
  plt.subplot(121),plt.imshow(image,'gray'),plt.title('Original')
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(dst,'gray'),plt.title('Barlett')
  plt.xticks([]), plt.yticks([])
  plt.show()
    
  return ()

def gaussiano(image):
  
  import numpy as np
  import cv2 as cv
  from matplotlib import pyplot as plt
  kernel = np.array ([[1, 4, 6,4,1],     # Este es el filtro establecido, 
                [ 4, 16, 24,16,4],           #que es el núcleo de convolución,funciona para mas dimensiones
                [ 6, 24, 36,24,6],
                [ 4, 16, 24,16,4],
                [ 1, 4, 6, 4, 1]])
  
  dst = cv.filter2D(image,-1,kernel)
  plt.subplot(121),plt.imshow(image,'gray'),plt.title('Original')
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(dst,'gray'),plt.title('Gaussiano')
  plt.xticks([]), plt.yticks([])
  plt.show()

 
    
  return ()
  
def  laplace4(image):
  import numpy as np
  import cv2 as cv
  from matplotlib import pyplot as plt
  kernel = np.array ([[0, -1, 0],     # Este es el filtro establecido, 
                [ -1, 4, -1],           #que es el núcleo de convolución,funciona para mas dimensiones
                [  0, -1, 0]])
  dst = cv.filter2D(image,-1,kernel) #cv lo normaliza
  plt.subplot(121),plt.imshow(image,'gray'),plt.title('Original')
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(dst,'gray'),plt.title('Laplace 4 vecinos')
  plt.xticks([]), plt.yticks([])
  plt.show()
  

  return ()


def laplace8(image):
  import numpy as np
  import cv2 as cv
  from matplotlib import pyplot as plt
  kernel = np.array ([[0, -1, 0],     # Este es el filtro establecido, 
                [ -1, 4, -1],           #que es el núcleo de convolución,funciona para mas dimensiones
                [  0, -1, 0]])
  dst = cv.filter2D(image,-1,kernel) #cv lo normaliza
  plt.subplot(121),plt.imshow(image,'gray'),plt.title('Original')
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(dst,'gray'),plt.title('Laplace 8 vecinos')
  plt.xticks([]), plt.yticks([])
  plt.show()
  
  
  return ()

def sobel(image):
  import numpy as np
  import cv2 as cv
  from matplotlib import pyplot as plt
  #orientación 1
  kernel = np.array ([[-1, 0, 1],     # Este es el filtro establecido, 
                [ -2, 0, +2],           #que es el núcleo de convolución,funciona para mas dimensiones
                [  -1, 0, 1]])
  dst = cv.filter2D(image,-1,kernel) #cv lo normaliza
  plt.subplot(121),plt.imshow(image,'gray'),plt.title('Original')
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(dst,'gray'),plt.title('Sobel 1')
  plt.xticks([]), plt.yticks([])
  plt.show()
  #orientación 2
  kernel2 = kernel.T
  dst = cv.filter2D(image,-1,kernel2) #cv lo normaliza
  plt.subplot(121),plt.imshow(image,'gray'),plt.title('Original')
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(dst,'gray'),plt.title('Sobel 2')
  plt.xticks([]), plt.yticks([])
  plt.show()
  #orientación 3
  kernel3 = kernel2.T
  dst = cv.filter2D(image,-1,kernel3) #cv lo normaliza
  plt.subplot(121),plt.imshow(image,'gray'),plt.title('Original')
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(dst,'gray'),plt.title('Sobel 3')
  plt.xticks([]), plt.yticks([])
  plt.show()
  #orientación 4
  kernel4 = kernel3.T
  dst = cv.filter2D(image,-1,kernel4) #cv lo normaliza
  plt.subplot(121),plt.imshow(image,'gray'),plt.title('Original')
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(dst,'gray'),plt.title('Sobel 3')
  plt.xticks([]), plt.yticks([])
  plt.show()
    
  return ()

def  pasabanda02(image):
  import cv2 as cv
  import numpy as np
  from matplotlib import pyplot as plt
  kernel = np.array ([[0, -1, 0],     # Este es el filtro establecido, 
                [ -1, 4, -1],           #que es el núcleo de convolución,funciona para mas dimensiones
                [  0, -1, 0]])
  identidad=np.identity(3)
  kernel=(kernel*0.2)+identidad

  dst = cv.filter2D(image,-1,kernel) #cv lo normaliza
  plt.subplot(121),plt.imshow(image,'gray'),plt.title('Original')
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(dst,'gray'),plt.title('Pasabanda 0.2')
  plt.xticks([]), plt.yticks([])
  plt.show()
  
  return ()


def  pasabanda04(image):
  import cv2 as cv
  import numpy as np
  from matplotlib import pyplot as plt
  kernel = np.array ([[0, -1, 0],     # Este es el filtro establecido, 
                [ -1, 4, -1],           #que es el núcleo de convolución,funciona para mas dimensiones
                [  0, -1, 0]])
  identidad=np.identity(3)
  kernel=(kernel*0.4)+identidad

  dst = cv.filter2D(image,-1,kernel) #cv lo normaliza
  plt.subplot(121),plt.imshow(image,'gray'),plt.title('Original')
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(dst,'gray'),plt.title('Pasabanda 0.4')
  plt.xticks([]), plt.yticks([])
  plt.show()
  
  return ()

#*******************************************************
#******************************************************
#***********  PROGRAMA PRINCIPAL  ***********************
#
def main():
  import matplotlib.pyplot as plt
  import pylab
  import cv2
  import numpy as np
  import imageio
  # programa principal
  ######################
  img = imageio.imread('imageio:camera.png') 
# kkernel = np.ones((5,5),np.float32)/25

  #************************
  #********MENU********
  salir = False
  opcion = 0
  while not salir:

    print ("1. Barlett")
    print ("2. Gaussiano")
    print ("3. Laplace 4 vecinos")
    print ("4. Laplace 8 vecinos")
    print ("5. Sobel ")
    print ("6. Pasabanda, pasaaltosde frecuencia de corte 0,2")
    print ("7. Pasabanda, pasaaltosde frecuencia de corte 0,4")
    print ("8 Salir")

    print('') 
    print ("Elige una opción") 
    opcion = int(input())
     
    if opcion == 1:
        barlett(img)
    elif opcion == 2:
        gaussiano(img)
    elif opcion == 3:
        laplace4(img)
    elif opcion == 4:
        laplace8 (img)
    elif opcion == 5:
        sobel(img)
    elif opcion == 6:
        pasabanda02(img)
    elif opcion == 7:
        pasabanda04(img)
    elif opcion == 8:
        salir = True
    else:
        print ("Introduce un número entre 1 y 5")
 
    
if __name__ == '__main__':
  main()
```
