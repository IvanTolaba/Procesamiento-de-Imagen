Modificaciones a la imagen

La imagen en RGB tiene tres bandas de colores

image[0] banda roja(R);  
image[1] banda verde(G);   
image[2] banda azul(B)  

```python
print(image.shape)           #dimension de la imagen
print(image[0].shape)        #dimension de la banda roja
print(image[1].shape)        #dimension de la banda verde
print(image[2].shape)        #dimension de la banda azul
```

Podemos ver la informacin de cada canal quitando la informacin de los otros canales. Por ejemplo, para ver el canal rojo, ponemos en 0 los canales azul y verde.

```python
image[:,:,1]=0     # Hago que valga 0 la banda G(verde)
image[:,:,2]=0     # Hago que valga 0 la banda B(azul)
print(image)       # La imagen al igual que las bandas valdran 0 en lo que correpsonda a verde y azul
```
